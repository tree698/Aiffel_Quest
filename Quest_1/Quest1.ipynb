```python
import random
from datetime import datetime

class BankAccount:
    bank_name = "SC은행"

    def __init__(self, name, balance):
        self.name = name
        self.balance = float(balance)
        self.account_number = f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"
        self.deposit_history = {}
        self.withdraw_history = {}
        self.deposit_count = 0

    def deposit(self, amount):
        if amount >= 1:
            self.balance += float(amount)
            time = datetime.now()
            self.deposit_count += 1
            self.deposit_history[time] = amount
            if self.deposit_count % 5 == 0:
                self.balance = round(self.balance * 1.01, 2)
                interest = self.balance * 0.01
                print(f'{interest:,.0f}원이 이자가 생겼습니다.')
            print(f'{amount:,.0f}원이 입금되었습니다.')
        else:
            print("입금은 1원 이상부터 가능합니다.")

    def withdraw(self, amount):
        if amount > self.balance:
            print('잔액이 부족합니다.')
        else:
            self.balance -= float(amount)
            time = datetime.now()
            self.withdraw_history[time] = amount
        print(f'{amount:,.0f}원이 출금되었습니다.')

    def display_info(self):
        balance_mod = f"{self.balance:,.0f}"
        print(f'은행이름: {BankAccount.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {balance_mod}원')

    def show_deposit_history(self):
        for key, val in self.deposit_history.items():
            print(f'{key}에 {val:,.0f}원을 입금하셨습니다.')

    def show_withdraw_history(self):
        for key, val in self.withdraw_history.items():
            print(f'{key}에 {val:,.0f}원을 출금하셨습니다.')

    def all_history(self):
        self.show_deposit_history()
        self.show_withdraw_history()

    def show_balance(self):
      print(f'잔액은 {self.balance:,.0f}원 입니다.')

def play():
    print('지금부터 계좌를 만들겠습니다.')
    user_name = input("이름을 알려주세요: ")
    user_balance = input("얼마를 입금하시나요?(최소금액은 1원 이상입니다): ")
    account = BankAccount(user_name, user_balance)
    print("계좌가 생성되었습니다.")

    while True:
        user = int(input('원하시는 번호를 입력하세요: (1: 입금, 2. 출금, 3. 정보출력, 4. 예금내역, 5. 출금내역, 6. 모든내역, 7. 잔고조회, 0. 종료): '))

        if user == 0:
            print('다음에 또 봐요!')
            return  # 프로그램 종료
        elif user == 1:
            while True:
                try:
                    amount = float(input('얼마를 입금하시나요? (최소 입금은 1원입니다): '))
                    if amount < 1:
                        raise ValueError()
                except ValueError:
                    print("잘못된 금액입니다. 다시 시도해 주세요.")
                else:
                    account.deposit(amount)
                    break
        elif user == 2:
            while True:
                try:
                    amount = float(input('얼마를 출금하시나요?: '))
                    if amount < 1:
                        raise ValueError()
                except ValueError:
                    print("잘못된 금액입니다. 다시 시도해 주세요.")
                else:
                    account.withdraw(amount)
                    break
        elif user == 3:
            account.display_info()
        elif user == 4:
            account.show_deposit_history()
        elif user == 5:
            account.show_withdraw_history()
        elif user == 6:
            account.all_history()
        elif user == 7:
            account.show_balance()
        else:
          print('잘못된 번호를 입력했습니다')

if __name__ == "__main__":
  play()
```
